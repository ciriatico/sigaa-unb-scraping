```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import Select

from bs4 import BeautifulSoup
import pandas as pd
```

```python
DRIVER_PATH = './selenium/chromedriver'
ser = Service(DRIVER_PATH)
op = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=ser, options=op)
```

```python
driver.get("https://sigaa.unb.br/sigaa/public/curso/lista.jsf?nivel=G&aba=p-ensino")
```

```python
main_page = driver.page_source
```

```python
main_soup = BeautifulSoup(main_page, 'html.parser')
```

```python
departamentos = main_soup.find_all('td', {'class': 'subFormulario'})
```

```python
cursos_par = main_soup.find_all('tr', {'class': 'linhaPar'})
cursos_impar = main_soup.find_all('tr', {'class': 'linhaImpar'})
cursos = sorted(cursos_par + cursos_impar, key=lambda x: x.sourceline)
```

```python
departamentos_cursos = []

for i in range(0, len(departamentos)):
    departamentos_cursos.append([departamentos[i], []])

    pos_departamento = departamentos[i].sourceline

    if i != (len(departamentos)-1):
        pos_prox_departamento = departamentos[i+1].sourceline
    else:
        pos_prox_departamento = None

    if pos_prox_departamento != None:
        for curso in cursos:
            if (curso.sourceline > pos_departamento) and (curso.sourceline < pos_prox_departamento):
                departamentos_cursos[i][1].append(curso)
    else:
        for curso in cursos:
            if (curso.sourceline > pos_departamento):
                departamentos_cursos[i][1].append(curso)
```

```python
base_url = 'https://sigaa.unb.br/sigaa/public/curso/'

cursos_dict = {
    'departamento': [],
    'curso': [],
    'grau': [],
    'turno': [],
    'sede': [],
    'modalidade': [],
    'coordenador': [],
    'url': []
}

for departamento_cursos in departamentos_cursos:    
    departamento = departamento_cursos[0].text.strip()
    
    for curso in departamento_cursos[1]:
        curso_infos = curso.find_all('td', {'class': None})
        curso_infos = [info.text.strip() for info in curso_infos]
        
        cursos_dict['departamento'].append(departamento)
        cursos_dict['curso'].append(curso_infos[0])
        cursos_dict['grau'].append(curso_infos[1])
        cursos_dict['turno'].append(curso_infos[2])
        cursos_dict['sede'].append(curso_infos[3])
        cursos_dict['modalidade'].append(curso_infos[4])
        cursos_dict['coordenador'].append(curso_infos[5])
        
        cursos_dict['url'].append(base_url + curso.find('a', {'class': None})['href'])
```

```python
cursos_df = pd.DataFrame(cursos_dict)
```
